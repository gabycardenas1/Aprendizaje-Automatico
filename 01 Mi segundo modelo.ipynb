## <font color='264CC7'> Introducción </font>

Este notebook está diseñado parcticar en el entrenamiento y evaluación de un modelo de regresión lineal utilizando datos reales. Los objetivos principales son:

- Entrenar un modelo de regresión lineal.
- Evaluar el ajuste del modelo.


Del conjunto de datos disponible [aquí](https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv), se desea predecir la variable `price` en función de las variables `bedrooms`, `bathrooms`, `sqft_living`, `sqft_lot`, `floors`.

```python
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler, KBinsDiscretizer, OneHotEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

---
## <font color='264CC7'> Lectura de Datos </font>


Realiza la lectura de los datos y presenta las primeras filas del conjunto de datos.

```python
df = pd.read_csv("https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv")
X = df[["bedrooms", "bathrooms", "sqft_living", "sqft_lot", "floors"]]
y = df["price"]
df_final = pd.concat([X, y.rename("price")], axis=1)

df_final.head()
```

---
## <font color='264CC7'> Preprocesamiento de datos </font>

Realiza una normalización de los datos, prueba diferentes técnicas y selecciona la que mejor se ajuste a los datos.

```python
# Normalización por el máximo
x_max = X.max()
X_max_normalized = X / x_max
print("Normalización por el máximo:\n")
display(X_max_normalized.head())

# Normalización Min-Max
min_max_scaler = MinMaxScaler()
X_min_max_normalized = pd.DataFrame(min_max_scaler.fit_transform(X), columns=X.columns)
print("Normalización Min-Max:\n")
display(X_min_max_normalized.head())

# Estandarización
standard_scaler = StandardScaler()
X_standardized = pd.DataFrame(standard_scaler.fit_transform(X), columns=X.columns)
print("Estandarización:\n")
display(X_standardized.head())
```

---
## <font color='264CC7'> Características y etiquetas </font>

Separa las características y las etiquetas y mostrar la dimensión de cada conjunto.

```python
# Dividir el conjunto de datos en características y etiqueta

features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']

X = df.loc[:, features]
y = df.loc[:, ['price']]
```

---
## <font color='264CC7'> Conjunto de entrenamiento y prueba </font>

Divide los datos en dos conjuntos: uno para entrenamiento y otro para prueba y visualiza la dimensión de cada conjunto. Usar `random_state` igual al últmo número de tu cédula.

```python
from sklearn.model_selection import train_test_split

# Divide los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

# Mostrar dimensiones de cada conjunto
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

Revisa la distribución de las etiquetas en los conjuntos de entrenamiento y prueba.

```python
print("Distribución en train:")
print(y_train.describe())

print("\nDistribución en test:")
print(y_test.describe())
```

---
## <font color='264CC7'> Entrenamiento </font>

Define un modelo de regresión lineal y lo entrena con los datos de entrenamiento.

```python
from sklearn.linear_model import LinearRegression

modelo_lineal = LinearRegression()
modelo_lineal.fit(X_train, y_train)


# Mostrar la ecuación completa del modelo lineal
intercepto = modelo_lineal.intercept_[0]
coeficientes = modelo_lineal.coef_[0]

print("Ecuación del modelo de regresión lineal:")
print(f"price = {intercepto:.4f} "
      f"+ ({coeficientes[0]:.4f})*bedrooms "
      f"+ ({coeficientes[1]:.4f})*bathrooms "
      f"+ ({coeficientes[2]:.4f})*sqft_living "
      f"+ ({coeficientes[3]:.4f})*sqft_lot "
      f"+ ({coeficientes[4]:.4f})*floors")
```

---
## <font color='264CC7'> Evaluacion </font>

Realiza la predicciones con el conjunto de prueba y muestra los valores en una tabla.

```python
# Realizar predicciones con el conjunto de prueba
y_pred = modelo_lineal.predict(X_test)

# Crear tabla con valores reales vs predicciones
df_predicciones = pd.DataFrame({'Datos reales': y_test.values.flatten(),'Predicciones': y_pred.flatten()})

# Mostrar tabla
df_predicciones.head()
```

Calcula el error cuadrático medio.

```python
from sklearn.metrics import mean_squared_error

# Cálculo del MSE
mse = mean_squared_error(y_test, y_pred)

print("Error cuadrático medio (MSE):", mse)
```

Escribe el error cuadrático medio para cada tipo de normalización:

- Normalización MaxAbsScaler:
- Normalización MinMaxScaler:
- Normalización RobustScaler:
- Normalización StandardScaler:
- Normalización Normalizer:
- Normalización QuantileTransformer:
- Normalización PowerTransformer:

```python
from sklearn.preprocessing import (MaxAbsScaler, MinMaxScaler, RobustScaler, StandardScaler, Normalizer, QuantileTransformer, PowerTransformer)

normalizadores = {
    'MaxAbsScaler': MaxAbsScaler(),
    'MinMaxScaler': MinMaxScaler(),
    'RobustScaler': RobustScaler(),
    'StandardScaler': StandardScaler(),
    'Normalizer': Normalizer(),
    'QuantileTransformer': QuantileTransformer(output_distribution='normal'),
    'PowerTransformer': PowerTransformer()
}

resultados = []

for nombre, scaler in normalizadores.items():

    # Ajustar normalizador solo con datos de entrenamiento
    scaler.fit(X_train)

    # Transformar train y test
    X_train_scaled = scaler.transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Crear modelo lineal
    modelo = LinearRegression()

    # Entrenar
    modelo.fit(X_train_scaled, y_train)

    # Predecir
    y_pred = modelo.predict(X_test_scaled)

    # MSE
    mse = mean_squared_error(y_test, y_pred)

    # Guardar
    resultados.append([nombre, mse])

df_resultados = pd.DataFrame(resultados, columns=['Normalización', 'MSE'])

df_resultados
```

Escribe un párrafo sobre qué hace normalización no vista en clase:

- **Normalización ...:** ...

**Normalización MaxAbsScaler:** Escala cada columna dividiéndola para su valor absoluto más grande, dejando los datos entre -1 y 1 sin alterar la forma de la distribución. Es útil cuando hay valores positivos y negativos y no queremos modificar la estructura original.

**Normalización RobustScaler:** Utiliza la mediana y el rango intercuartil para escalar los datos, por lo que resiste muy bien la presencia de outliers. Es ideal cuando existen valores extremos que podrían afectar métodos más sensibles.

**Normalización Normalizer:** A diferencia de otros métodos, normaliza cada fila completa convirtiéndola en un vector de longitud 1. Se usa cuando importa más la proporción entre características que sus magnitudes, como en análisis basados en similitud.

**Normalización QuantileTransformer:** Transforma los datos asignándoles nuevos cuantiles para que sigan una distribución más uniforme o normal. Es práctico cuando la variable tiene mucha asimetría o concentración en ciertos rangos.

**Normalización PowerTransformer:** Aplica una transformación matemática llamada Box-Cox o Yeo-Johnson para reducir el sesgo y estabilizar la varianza. Ayuda a que la variable se acerque a una distribución normal, mejorando el desempeño en modelos lineales.
